```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/diabetes.csv")
```

```python
df.info(verbose=True, null_counts=True)  #No NULL Values
```

```python
df
```

```python
x = df.iloc[:,:8].values
y = df.iloc[:,8].values
```

```python
y #Only 0's and 1's
```

```python
df.corr() #No correlation
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
Sc = StandardScaler()
```

```python
x_t_train = Sc.fit_transform(x_train)  #Standard Scaling Dataset
```

```python
x_t_train
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression #Implementing Logistic Regression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(x_t_train,y_train)
```

```python
pred_LR = LR.predict(Sc.transform(x_test))
pred_LR
```

```python
y_test
```

```python
accuracy_score(pred_LR,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier #Implementing KNN
```

```python
KNN = KNeighborsClassifier(n_neighbors=20)
KNN.fit(x_t_train,y_train)
```

```python
pred_KNN = KNN.predict(Sc.transform(x_test))
pred_KNN
```

```python
y_test
```

```python
accuracy_score(pred_KNN,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier #Implementing Decision Tree
```

```python
DT = DecisionTreeClassifier()
```

```python
DT.fit(x_t_train,y_train)
```

```python
pred_DT = DT.predict(Sc.transform(x_test))
pred_DT
```

```python
y_test
```

```python
accuracy_score(pred_DT,y_test)
```

```python
from sklearn.svm import SVC #Implementing Support Vector Classifier
```

```python
svm = SVC()
```

```python
svm.fit(x_t_train,y_train)
```

```python
pred_svm = svm.predict(Sc.transform(x_test))
pred_svm
```

```python
y_test
```

```python
accuracy_score(pred_svm,y_test)
```
